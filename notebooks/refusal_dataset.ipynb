```python
from datasets import load_dataset

sapien_w_context = load_dataset("eve-esa/open_ended_w_context")
```

```python
import random

def get_document(sample):
    labels = ["Doc 1", "Doc 2", "Doc 3"]
    for label in labels:
        if sample[label] is not None:
            return sample[label]
    return ""

def randomize_context(split):
    n = len(split)
    context = [get_document(sample) for sample in split]

    indexes = list(range(n))
    random.shuffle(indexes)

    randomized_context = [context[i] for i in indexes]

    return randomized_context


for split in sapien_w_context:
    randomized_context = randomize_context(sapien_w_context[split])
    sapien_w_context[split] = sapien_w_context[split].remove_columns(['Doc 1', 'Doc 2', 'Doc 3']).add_column('Context', randomized_context)
```

```python
sapien_w_context.remove_columns('Source 1').remove_columns('Source 2').remove_columns('Source 3')
```

```python
# Push to hub
# Push to the hub
for split, dataset in sapien_w_context.items():
    dataset.push_to_hub("eve-esa/sapien_open_ended_refusal", split ,split='train', private=True)
```
